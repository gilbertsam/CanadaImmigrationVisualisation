<h1> Area Visualization </h1>

```python
df.sort_values(by='Total', ascending=False, axis=0, inplace=True)
df_top5=df.head(5)
df_top5=df_top5[years].transpose()
df_top5.index = df_top5.index.map(int)
df_top5.plot(kind='area',alpha=0.2,stacked=False, figsize=(20,10))
plt.title('Top 10 immigration countries to Canada')
plt.ylabel('Number of immigrants')
plt.xlabel('Year')
plt.show()
```

Pada grafik diatas, terjadi perubahaan negara mana yang menduduki jumlah imigran tertinggi pada tiap tahun seperti pada tahun 1980- 1995 diduduki oleh UK atau Inggris dan pada awal 2000-an diganti dengan negara China dan pada awal tahun 2010, diganti dengan banyaknya masyarakat filipina dan diraih balik oleh China. 

```python
df.sort_values(by='Total', ascending=False, axis=0, inplace=True)
df_top10=df.head(10)
df_top10=df_top10[years].transpose()
df_top10.index = df_top10.index.map(int)
df_top10.plot(kind='area',alpha=0.2,stacked=False, figsize=(20,10))
plt.title('Top 10 immigration countries to Canada')
plt.ylabel('Number of immigrants')
plt.xlabel('Year')
plt.show()
```

Pada grafik data diatas, dapat dilihat perbedaan tiga negara tersebut. Untuk negara tertinggi masih dimiliki oleh UK pada tahun 1980 - 1995 dan dilanjutin oleh China pada tahun 2000-2005 dan digantikan oleh filipina dan pada tahun 2005-2010 dan diganti lagi oleh Cina 

<h1> Histogram Visualization </h1

```python
#Histogram Visualisation
hist,bin_edges=np.histogram(df[2013])
print(hist)
print(bin_edges)

df[2013].plot(kind='hist',figsize=(10,5), xticks=bin_edges)
plt.title('Immigration from 195 Countries to Canada in 2013')
plt.ylabel('Number of countries')
plt.xlabel('Number of Immigrants')
plt.show()
```

Berdasarkan histogram, kita melihat kecondongan data pada tahun 2013 cenderung condong ke kiri. Untuk angka 3413 dimiliki oleh 175 negara. Dengan populasi diatas angkat 3413 dimiliki oleh beberapa negara kurang dari 25 negara.

```python
hist,bin_edges=np.histogram(df[1980])
print(hist)
print(bin_edges)

df[1980].plot(kind='hist',figsize=(10,5), xticks=bin_edges)
plt.title('Immigration from 195 Countries to Canada in 1980')
plt.ylabel('Number of countries')
plt.xlabel('Number of Immigrants')
plt.show()
```

Pada tahun 1980, memiliki histogram yang sama dengan pada tahun 2013 yakni condong ke kiri.

<h1> Scatter Plot Visualization </h1>

Untuk memulai visualisasi scatter plot perlu di inisiasikan terlebih dahulu sebuah variabel yang menyimpan data tahun dan total. 

```python
/#Inisiasikan dataframe total.
df_tot = pd.DataFrame(df[years].sum(axis=0))
df_tot.index = map(int, df_tot.index)
df_tot.reset_index(inplace = True)
df_tot.columns = ['year', 'total']
df_tot.head()
```

Setelah diinisiasikan, maka kita dapat membuat grafik dengan variable df_tot. Variabel x berupa tahun dan variabel y berupa total. 

```python
#Pembuatan grafik
df_tot.plot(kind='scatter', x='year', y='total', figsize=(10, 6), color='darkblue')

plt.title('Total Immigration to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')

plt.show()
```

Kalau dilihat tanpa mengunakan garis, maka dapat dilihat adanya bentuk secara linear hal ini kita coba dengan tarik garis dari titik (1980, +-10000) menuju tahun 2010 keatas dengan nilai y sekitar 25000. 

Setelah menghitung total, kita dapat menggunakan bentuk scatter plot untuk menghitung nilai rata-rata imigrant tiap tahun. 

```python
df_mean= pd.DataFrame(df[years].mean(axis=0))

#Membuat index
df_mean.index= map(int, df_mean.index)
df_mean.reset_index(inplace = True)

df_mean.columns = ['Year', 'Mean']

#Membuat grafik
df_mean.plot(kind='scatter', x='Year', y='Mean', figsize=(10, 6))

plt.title('Average Immigration to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')

plt.show()
```

Grafik diatas, terlihat sama seperti grafik sebelumnya yakni positif keatas dan dapat ditarik sebuah garis linear. Tujuan dari garis linear ini adalah untuk membantu sebuah pemodelan nanti seperti memprediksi jumlah imigran datang dengan menggunakan metode linear regresion.
